```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp
```

### 1. Membership of $f_a$ into the scale $C^{k,\alpha}([-1,1])$

We start by calculating $f_a^{(k)}$. We have 
$$
f_a'(x) = \begin{cases} ax^{a-1} &\text { for } x > 0\\ 0 &\text { for } x \le 0 \end{cases},
$$

$$
f_a''(x) = \begin{cases} a(a-1)x^{a-1}  &\text { for } x > 0\\ 0 &\text { for } x \le 0 \end{cases}, \quad (a \ne 1)
$$

therefore for any $k \in \mathbb N$ such that $a \not \in \{1, 2,3, \ldots, k - 1\}$, we have:

$$
f_a^{(k)}(x) = \begin{cases} a(a-1) \cdots (a-k+1)x^{a-k}  &\text { for } x > 0\\ 0 &\text { for } x \le 0 \end{cases}.
$$

Clearly $f_a^{(k)}$ is continuous separately on the intervals $[-1,0]$ and $(0, 1]$. We need to show that it is continuous at $x=0$.

* If $k = a$, then $f_a^{(k)}(x) = a(a-1) \cdots (a-k+1)$ for $x > 0$. Hence $\lim_{x \to 0^+} f_a^{(k)}(x) = a(a-1) \cdots (a-k+1) \ne 0$, so $f_a^{(k)} \not \in C^k([-1, 1])$.

* If $k > a$ and $a \not \in \{1,2,3,\ldots, k -1\}$, then $a - k < 0$, hence $\lim_{x \to 0^+} f_a^{(k)}(x) = \lim_{x \to 0^+} a(a-1) \cdots (a-k+1) x^{a-k} = \pm \infty \ne 0$. Therefore $f_a^{(k)} \not \in C^k([-1, 1])$.

* If $k < a$, then $a - k > 0$, and $\lim_{x \to 0^+} f_a^{(k)}(x) = \lim_{x \to 0^+} a(a-1) \cdots (a-k+1)x^{a-k} = 0$. Therefore $f_a^{(k)} \in C^k([-1, 1])$.

From this, it follows that we need to have $a - k < 0$.

-----------------------------------


Now we will show that we must have $k + \alpha \le a$, i.e. $\alpha \le a -k$. Suppose by contradiction that $k + \alpha > a$. Then for any $x \in (0, 1]$, we have:

$$
\frac {|f_a^{(k)}(x) - f_a^{(k)}(-x)|} {|x + x|^\alpha} = \frac {|a(a-1) \cdots (a-k+1) x^{a-k}|} {|2x|^\alpha} = \frac {|a(a-1) \cdots (a-k+1)|} {2^{\alpha}} \cdot x^{(a - k - \alpha)}.
$$

As $x \to 0^{+}$, the right hand side will go to $\pm \infty$ since $a - k - \alpha < 0$. Therefore for any $\alpha > a$, 

$$
\sup_{-1 \le x < y \le 1} \frac {|f^{(k)}_a(x) - f^{(k)}_a(y)|} {|x-y|^\alpha} = +\infty.
$$

Hence $f \not \in C^{0, \alpha}([-1, 1])$ for $k + \alpha > a$, so we must have $\alpha + k \le a$.


------------------------------------

Now we wish to show that $f_a^{(k)} \in C^{0,\alpha}([-1, 1])$, assuming that $\alpha \le a - k$ and $0 \le a - k \le 1$. Let $x, y \in \mathbb R$ such that $-1 \le x < y \le 1$.

* If $x, y \le 0$, then

$$
\frac {|f^{(k)}_a(x) - f^{(k)}_a(y)|} {|x-y|^{\alpha}} = 0
$$

* If $-1 \le x \le 0 < y \le 1$, then we have 

$$
\frac {|f^{(k)}_a(x) - f^{(k)}_a(y)|} {|x-y|^{\alpha}} = \frac {|a(a-1)\cdots (a-k+1)y^{a-k}|} {|x - y|^\alpha} \le \frac {|a(a-1) \cdots (a-k+1)|} {2^\alpha} y^{a - k}
$$
and the right hand side is bounded since $a -k \ge 0$.

* If $0 < x < y \le 1$, we will first let $\ell = y - x$. Then we have:

$$
\frac {|f^{(k)}_a(x) - f^{(k)}_a(y)|} {|x - y|^\alpha} = |a(a-1)\cdots (a-k+1)| \cdot \frac{y^{a - k} - x^{a-k}} {(y - x)^\alpha} = |a(a-1)\cdots (a-k+1)y^{a-k}| \cdot \frac{(x+\ell)^{a-k} - x^{a-k}} {\ell^{a-k}}
$$

Now, because $0 \le a -k \le 1$, we have the triangle inequality, i.e. $(x+\ell)^{a-k} \le x^{a-k} + \ell^{a-k}$ since both $x$ and $\ell$ are positive. Therefore, we have 

$$
\frac {|f^{(k)}_a(x) - f^{(k)}_a(y)|} {|x - y|^\alpha} \le |a(a-1)\cdots (a-k+1)| \cdot \frac{\ell^{a-k}} {\ell^{a-k}} = |a(a-1)\cdots (a-k+1)y^{a-k}|
$$

Therefore the right hand side is bounded. It follows that for $0 \le a - k \le 1$ and $k + \alpha \le a$, we have:
$$
\sup_{-1 \le x < y \le 1} \frac {|f_a^{(k)}(x) - f_a^{(k)}(y)|} {|x-y|^\alpha} \le \infty.
$$
thus $f_a \in C^{k,\alpha}([-1,1])$ for $0 < \alpha \le a - k \le 1$.

---------------------

Hence the largest $k+\alpha$ such that $f_a \in C^{k, \alpha}([-1,1])$ is $k+\alpha = a$.

### 2. Piecewise Constant Approximation
We start by defining a function `f_a` that takes as input $a > 0$ and returns the function $f_a$. We will use this function throughout the notebook.

We then define the function `interpolate_constant` to approximate $f_a$ by a piecewise constant function on $[-1, 1]$ with $n$ partitions. In order to approximate it by piecewise constant functions, for $x \in [-1, 1]$, we find the index of the corresponding interval by using the formula:

$$
    i = \begin{cases} \text{floor}\left ((x + 1) \cdot \frac {n} 2 \right) & x < 1 \\
    n - 1 & x = 1
    \end{cases}.
$$

We then plot the function $f_a$ and its interpolation for $a \in \{0.5, 1, 2\}$ and $n \in \{4, 9, 32\}$.

```python
def f_a(a):
    def f(x):
        if x <= 0: return 0
        else     : return x ** a
    return f
```

```python
def interpolate_constant(a, n):
    xs = np.linspace(-1, 1, n + 1)
    ys = [f_a(a)((xs[i] + xs[i+1]) / 2) for i in range(n)]
    return lambda x: ys[n-1] if x == 1 else ys[int(np.floor((x + 1) * n / 2))]

xs = np.linspace(-1, 0.99, 400)

for a in (0.5, 1.0, 2.0):
    plt.figure(figsize=(12, 6))
    plt.plot(xs, [f_a(a)(x) for x in xs], label=r'$f_{%.1f(x)}$' % a)
    plt.plot(xs, [interpolate_constant(a, 4)(x) for x in xs], label=r'$n=4$')
    plt.plot(xs, [interpolate_constant(a, 9)(x) for x in xs], label=r'$n=9$')
    plt.plot(xs, [interpolate_constant(a, 32)(x) for x in xs], label=r'$n=32$')
    plt.legend()
    plt.title(r"Graph of $f_{%.1f}(x)$ and its piecewise constant approximations for various $n$" % a)
    plt.show()
```

In order to compute the error, we define a general function that estimate the distance in the maximum norms between two functions defined on $[-1,1]$. We will reuse this function for the later parts.

```python
def max_error(f, g, n=1000):
    xs = np.linspace(-1, 1, n)
    return max([np.abs(f(x) - g(x)) for x in xs])

err = lambda a, n: max_error(f_a(a), interpolate_constant(a, n))
```

Now we can try to find the rate of convergence: we define the function `find_r` for this purpose, using `np.polyfit`. In order to verify that the output is correct and that small values of $n$ do not interfere with the result, we start by plotting the error for $a=0.8$ and different $n$, and also plot $e^r$ using  the rate of convergence $r$ we found. We see that the two curves match.

Now we can plot the rate of convergence versus $a$. The rate of convergence seems to reach its maximum at $a=1.0$. This could be expected, since when $a$ is close to $1$ the error is more evenly distributed. It seems to decrease at a constant rate for $a > 1$.

```python

def find_r(a):
    ns = range(200, 500)
    (_, r) = np.polyfit(deg=1, x=ns, y=[n ** (-err(a, n)) for n in ns])
    return r

def plot_test():
    a = 0.8
    r = find_r(a)
    ns = range(1, 200)
    plt.figure(figsize=(12, 6))
    plt.plot(ns, [err(a, n) for n in ns], label=r'error in the maximum norm')
    plt.plot(ns, [n ** (-r) for n in ns], label=r'$n^{-%.2f}$' % r)
    plt.legend()
    plt.title("Error in the maximum norm vs $n$ for $a$=0.8 and asymptotic rate of convergence")
    plt.show()
    
plot_test()

plt.figure(figsize=(15, 10))
plt.subplot(221)
as_ = np.linspace(0.5, 5, 50)
rs = [find_r(a) for a in as_]
plt.plot(as_, rs, '-o')
plt.title("Rate of convergence vs $a \in [0.5, 5]$")
plt.xlabel('$a$')
plt.ylabel('rate of convergence')

plt.subplot(222)
as_ = np.linspace(0.5, 1.5, 50)
rs = [find_r(a) for a in as_]
plt.plot(as_, rs, '-o')
plt.xlabel('$a$')
plt.ylabel('rate of convergence')
plt.title("Rate of convergence vs $a \in [0.5, 1.5]$")
plt.show()
```

### 3. Piecewise Linear Interpolation

We start by defining `interpolate_linear` to compute the interpolation of $f_a$ by continuous piecewise linear fuction on the uniform partion of $[-1, 1]$. We then plot the function $f_a$ and its interpolation for $a \in \{0.5, 1.2, 2\}$ and $n \in \{4, 8, 32\}$, along with the error in the maximum norm vs $n$ for each $a$.

```python
def interpolate_linear(a, n):
    ''' Interpolate $f_a$ by a continous piecewise linear function with n intervals '''
    xs = np.linspace(-1, 1, n+1)
    ys = [f_a(a)(x) for x in xs]
    def f(x):
        i = int(np.floor((x + 1) / 2 * n)) # interval on which x lies
        if i == n:
                return ys[i]
        return ys[i] + (ys[i+1] - ys[i]) * (x - xs[i]) * (n / 2)
    return f

# Error in the maximum norm
err = lambda a, n: max_error(f_a(a), interpolate_linear(a, n), n=1000)

for a in (1.7, 2.0, 2.3):
    plt.figure(figsize=(15,5))
    
    plt.subplot(121)
    xs = np.linspace(-1, 1, 200)
    plt.plot(xs, [f_a(a)(x) for x in xs], label='$f_{%.1f}(x)$' % a)
    for n in (4, 8, 32):
        interpolation = interpolate_linear(a, n)
        plt.plot(xs, [interpolation(x) for x in xs], label='interpolation for $n=%d$' % n)
    plt.title("Linear interpolation for $a=%.1f$" % a)
    plt.legend()
    
    plt.subplot(122)
    ns = range(50, 200)
    plt.plot(ns, [err(a, n) for n in ns])
    plt.title("Error in the maximum norm vs $n$ for $a=%.1f$" % a)
    plt.show()
```

 As before, the error of the approximation seems to behave like $n^{-r}$. We observe a distinct phenomenon in some cases, for example when $a=1.7$: the error is oscilatting. We guess that this is due to whether $n$ is odd or even, and whether $x=0$ is an interpolation point. However, this phenomenon becomes insignificant when $n$ is large enough, so we will use $n > 200$ to ensure that the rate of convergence we find is correct.

Below, we redefine `find_r`. We show two plots, one for $a \in [0.5, 5]$ and one centered around 2, for $a \in [1.8, 2.2]$. The rate of convergence looks similar to the one found for constant interpolation, but reaches its maximal at $a=2$. We ca It also looks to decrease at a constant rate for $a > 2$.

```python
def find_r(a):
    ns = range(200, 300)
    (_, r) = np.polyfit(deg=1, x=ns, y=[n ** (-err(a, n)) for n in ns])
    return r

plt.figure(figsize=(15, 10))
plt.subplot(221)
as_ = np.linspace(0.5, 5, 50)
rs = [find_r(a) for a in as_]
plt.plot(as_, rs, '-o')
plt.xlabel('$a$')
plt.ylabel('rate of convergence')
plt.title("Rate of convergence vs $a \in [0.5, 5]$ using piecewise constant interpolation")

plt.subplot(222)
as_ = np.linspace(1.8, 2.2, 50)
rs = [find_r(a) for a in as_]
plt.plot(as_, rs, '-o')
plt.xlabel('$a$')
plt.ylabel('rate of convergence')
plt.title("Rate of convergence vs $a \in [1.8, 2]$ using piecewise constant interpolation")
plt.show()
```

### 4. Bernstein Approximation
We first define a function to compute the Bernstein basis polynomials on $[0,1]$. We then use this to define the function `bernsteins_approximate`, which returns a $n$-th degree Bernstein approximation to $f_a$.

We proceed to plot $f_a$ and its approximations for a few $n$, along with the error in the maximum norm vs $n$ for the same $a$.

```python
def bernstein_basis(x, n):
    ''' Compute the n+1 n-th degree Bernstein basis polynomials evaluated at x'''
    i = np.arange(n+1)
    binoms = sp.binom(n, i)
    return binoms * (x ** i) * ((1. - x) ** (n - i))

def bernsteins_approximate(a, n):
    ''' Compute the n-th degree Bernsteins approximation of $f_a$ '''
    f = f_a(a)
    def g(x):
        x = (x + 1) / 2
        bs = bernstein_basis(x, n)
        return np.sum(bs * np.array([f((i / n) * 2 - 1) for i in range(n+1)]))
    return g

err = lambda a, n: max_error(bernsteins_approximate(a, n), f_a(a), n=500)

for a in (0.5, 1., 1.5, 2., 2.5):
    xs = np.linspace(-1, 1, 300)
    plt.figure(figsize=(15, 5))
    
    plt.subplot(121)
    plt.plot(xs, [f_a(a)(x) for x in xs], label='$f_{%.1f}(x)$' % a)
    plt.plot(xs, [bernsteins_approximate(a, 4)(x) for x in xs], label='$n=4$')
    plt.plot(xs, [bernsteins_approximate(a, 8)(x) for x in xs], label='$n=8$')
    plt.plot(xs, [bernsteins_approximate(a, 32)(x) for x in xs], label='$n=32$')
    plt.title('Bernstein Approximation for $a=%.1f$' % a)
    plt.legend()
    
    plt.subplot(122)
    ns = range(10, 600)
    plt.plot(ns, [err(a, n) for n in ns])
    plt.title('Error in the max norm vs $n$ for $a=%.1f$' % a)
    plt.show()
    
```

The error seems to behave like $n^{-r}$ again.  Note that we have the "oscillation" phenomenon again for $a=0.5$,  which is why the curve looks thick. 

Now we wish to estimate the rate of convergence $r$ for different values of $a$. Since the implementation of Bernstein approximation above is a bit slow, we will not use the same method as before (using `np.polyfit`). Instead, we know that the error $e_n$ is given by $e_n \approx n^{-r}$. So we must have $$
r \approx \frac {\log n} {\log e_n} 
$$
so we will find $r$ by using this formula on $n$ large enough.

We observe that the rate of convergence looks to reach its maximum at $x \approx 2.1$, taking a value close to $1$.

```python
def find_r(a):
    n = 600
    return -np.log(err(a, n)) / np.log(n)

plt.figure(figsize=(15, 10))
plt.subplot(221)
as_ = np.linspace(0.5, 5.0, 64)
rs = [find_r(a) for a in as_]
plt.plot(as_, rs, '-o')
plt.title("Rate of convergence vs $a \in [0.5, 5]$ using Bernsteins approximation")
plt.xlabel('$a$')
plt.ylabel('rate of convergence')

plt.subplot(222)
as_ = np.linspace(1.7, 2.3, 64)
rs = [find_r(a) for a in as_]
plt.plot(as_, rs, '-o')
plt.xlabel('$a$')
plt.ylabel('rate of convergence')
plt.title("Rate of convergence vs $a \in [1.7, 2.3]$ using Bernsteins approximation")
plt.show()
```

### 5. Chebyschev Interpolation

We start by defining a function that evaluate $T_n$ at a point $x \in [-1, 1]$. We then define `chebyschev_interpolate`, which returns the $n$-th degree Chebyschev interpolation for $f_a$. As before, we then plot the function and its approximation for several $n$'s, and the error in the maximum norm for the same $a$'s.

```python
def chebyschev_interpolate(a, n):
    f = f_a(a)
    xs = np.cos((np.arange(0, n+1) + 0.5) * np.pi / (n+1)) # Chebyschev nodes
    ys = np.vectorize(f)(xs)
    s =  [np.cos((np.arange(0, n+1) + 0.5) * np.pi / (n+1) * i) for i in range (n+1)]
    coeffs = np.array([np.sum(s[i] * ys) for i in range(n+1)]) * 2/(1+n)
    def p(x):
        return np.sum(coeffs * np.cos(np.arange(0, n+1) * np.arccos(x))) - 0.5 * coeffs[0]
    return p

err = lambda a, n: max_error(chebyschev_interpolate(a, n), f_a(a), 500)
    
for a in (0.8, 1.2, 2.0, 2.5):
    xs = np.linspace(-1, 1, 1000)
    plt.figure(figsize=(15, 5))
    plt.subplot(121)
    plt.plot(xs, [f_a(a)(x) for x in xs])
    for n in (4, 8, 32):
        interpolation = chebyschev_interpolate(a, n)
        plt.plot(xs, [interpolation(x) for x in xs], label='$n = %d$' % n)
    plt.title('Chebyschev Interpolation for $a=%.1f$' % a)
    plt.legend()
    
    plt.subplot(122)
    ns = range(10, 400)
    plt.plot(ns, [err(a, n) for n in ns])
    plt.title('Error in the max norm vs $n$, for $a=%.1f$' % a)
    plt.show()
```

Once again, the error behaves like $n^{-r}$ for some $r$.

After estimating the rate of convergence, we observe that for $a \ge 4$, the rate of convergence seems to become constant. However, by changing the value of $n$ used to compute the rate of convergence, this constant part moves to the right or the left, so we guess that it is due to the way we compute the error. We also observe that the rate of convergence seems to increase linearly up until $a=2$, then keep increasing linearly but more slowly.

```python
def find_r(a):
    n = 2000
    return -(np.log(err(a, n)) / np.log(n))

plt.figure(figsize=(15, 10))
plt.subplot(221)
as_ = np.linspace(0.5, 5.0, 64)
rs = [find_r(a) for a in as_]
plt.plot(as_, rs, '-o')
plt.title("Rate of convergence vs $a \in [0.5, 5]$ using Chebyschev interpolation")
plt.xlabel('$a$')
plt.ylabel('rate of convergence')

plt.subplot(222)
as_ = np.linspace(1.95, 2.05, 64)
rs = [find_r(a) for a in as_]
plt.plot(as_, rs, '-o')
plt.xlabel('$a$')
plt.ylabel('rate of convergence')
plt.title("Rate of convergence vs $a \in [1.95, 2.05]$ using Chebyschev interpolation")
plt.show()
```
